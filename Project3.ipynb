## 데이터 전처리

#### 'start_date' 처리하기

```python
data['start_date'].unique()
```

```python
data.drop('start_date', axis=1, inplace=True)
```

```python
data.head()
```

#### 'location' 처리하기

```python
data['location'].unique()
```

```python
data = data.assign(location=data['location'].str.split(',')).explode('location').reset_index(drop=True)
```

```python
data
```

#### 'experience' 처리하기

```python
data['experience'].unique()
```

```python
# experience 처리
data = data.assign(
    min_experience=data['experience'].str.split(" ").str[0].str.split("-").str[0].astype("Int64"),
    max_experience=data['experience'].str.split(" ").str[0].str.split("-").str[-1].astype("Int64")
).drop(columns=['experience'])
```

```python
data['min_experience'].unique()
```

```python
data.head()
```

#### 'posted' 처리하기

```python
data['posted'].unique()
```

```python
def preprocess_posted(posted):
    posted = posted.split(" ")
    if posted[0] == "Just":
        return 1
    elif posted[0] == "Few":
        return 12
    elif posted[0] == "Today":
        return 24
    elif len(posted) == 3:
        if posted[1] == 'days' or posted[1] == 'day':
            return (int(posted[0]) + 1) * 1 * 24
        elif posted[1] == 'weeks' or posted[1] == 'week':
            return (int(posted[0]) + 1) * 7 * 24
```

```python
data['posted'] = data['posted'].str.replace("\n\n\nBe an early applicant", "")
data['posted'] = data['posted'].apply(preprocess_posted)
```

```python
data['posted'].unique()
```

```python
data.head()
```

#### 'ctc' 처리하기

```python
data['ctc'].unique()
```

```python
# ctc 처리
from currency_converter import CurrencyConverter

c = CurrencyConverter()

def calculate_ctc(ctc):
    ctc = ctc.split(" ")
    min , max = None, None

    if ctc[0] == "₹":
        min = c.convert(ctc[1].replace(",", ""), 'INR', 'USD')
        max = c.convert(ctc[-1].replace(",", ""), 'INR', 'USD')
    elif ctc[0] == "$":
        min = float(ctc[1].replace(",", ""))
        max = float(ctc[-1].replace(",", ""))
        
    return min, max

data[['min_ctc', 'max_ctc']] = data['ctc'].apply(calculate_ctc).apply(pd.Series)
data['avg_ctc'] = (data['min_ctc'] + data['max_ctc']) / 2
data.drop('ctc', axis=1, inplace=True)
```

```python
data.head()
```

## 회사가 많이 있는 도시를 내림차순으로 정렬해서 plotting 하시요.

```python
# 도시별 회사 수 카운트
city_counts = data['location'].value_counts().head(20)

# 도시별 데이터 시각화
plt.figure(figsize=(12, 6))
city_counts.plot(kind='bar')
plt.title('Top 20 Cities with Most Companies')
plt.xlabel('City')
plt.ylabel('Number of Companies')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 평균 임금 Best 10 & Worst 10 직종 출력

```python
# 직종별 평균 임금 계산
job_avg_ctc = data.groupby('job_title')['avg_ctc'].mean().dropna().sort_values()

# 상위 10개 직종
best_jobs = job_avg_ctc.tail(10)
print("Best 10 Jobs by Average Salary:")
print(best_jobs)

# 하위 10개 직종
worst_jobs = job_avg_ctc.head(10)
print("\nWorst 10 Jobs by Average Salary:")
print(worst_jobs)
```

## 경력별 평균 임금 출력

```python
# 경력별 평균 임금 계산
experience_avg_ctc = data.groupby('min_experience')['avg_ctc'].mean().dropna()

# 출력
print("Average Salary by Experience:")
print(experience_avg_ctc)
```

## JobFinder 클래스 구현

```python
class JobFinder:
    def __init__(self, salary, experience):
        self.salary = salary
        self.experience = experience

    def get_appropriate_job(self, n):
        recommended_jobs = data[
            (data['min_ctc'] >= self.salary) & 
            (data['min_experience'] <= self.experience) & 
            (data['max_experience'] >= self.experience)
        ].head(n)
        return recommended_jobs

    def print_info(self):
        print(f"Salary Expectation: {self.salary}")
        print(f"Experience: {self.experience} years")
```
