```python
# start_date 제거
data.drop('start_date', axis=1, inplace=True)
```

```python
# experience 처리
data['experience'] = data['experience'].str.replace(" years", "")
data['min_experience'] = data['experience'].str.split("-").str[0]
data['max_experience'] = data['experience'].str.split("-").str[-1]
data.drop('experience', axis=1, inplace=True)
```

```python
# ctc 처리
from currency_converter import CurrencyConverter
c = CurrencyConverter()

def calculate_ctc(ctc):
    list = ctc.split(" ")
    min , max = 0, 0

    if list[0] == "₹":
        min = c.convert(list[1].replace(",", ""), 'INR', 'USD')
        max = c.convert(list[-1].replace(",", ""), 'INR', 'USD')
    elif list[0] == "$":
        min = list[1]
        max = list[-1]
    else:
        print(list[0])

    return min, max


data['ctc'] = data['ctc'].apply(calculate_ctc)
```
